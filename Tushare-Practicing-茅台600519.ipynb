假如我从2010年1月1日开始，每月第一个交易日买入一手股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？

```python
price_last = df['open'][-1]
df = df ['2001-09':'2020-12']#剔除收尾无用数据
```

```python
df
```

```python
df.resample("M").sum()
```

```python
df.resample("MS").last()
```

```python
df_monthly = df.resample("MS").first()
```

```python
df_yearly = df.resample("A").last()[:-1]
```

```python
cost_money = 0
hold = 0
for year in range (2001,2020):
    
    cost_money += df_monthly[str(year)]['open'].sum()*100
    hold += len(df_monthly[str(year)]['open'])*100
    if year != 2017:
        cost_money -= df_yearly[str(year)]['open'][0]*hold
        hold = 0
    print(cost_money)
        
cost_money -= hold * price_last

print(-cost_money)
        

#df_monthly['2001']
```
